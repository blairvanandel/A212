```python
%matplotlib inline

xrange = np.arange(-np.pi,np.pi+0.01,0.01)
sinxrange = np.sin(xrange)
cosxrange = np.cos(xrange)

fig,ax1 = plt.subplots(1,1)
ax1.plot(xrange,sinxrange, color = "g", label='sin(x)')
ax1.plot(xrange,cosxrange, color = "k", label='cos(x)')
ax1.axhline(color="gray", y=0)
ax1.axvline(color="gray", x =0)
ax1.legend(loc = "best")
ax1.set(xlim=(-np.pi,np.pi))
ax1.set(title = "Cos(x) & Sin(x)")
```

```python
#read in datafile
t, d, dy = np.loadtxt("A212_day7data.txt", skiprows=5, unpack=True)
theoryt = np.arange(1,43.01,0.01)
theoryd = (3+((1/2)*np.sin((np.pi*theoryt)/5)))*theoryt*np.e**(-theoryt/10)

fig,ax2 = plt.subplots(1,1)
ax2.plot(theoryt,theoryd, color='gray', label='theory')
ax2.errorbar(t,d, yerr=dy, fmt='ro',ecolor='blue')
ax2.set(xlabel="time(s)",ylabel="position(cm)")
```

```python
d=np.array([0.38, 0.64, 0.91, 1.26, 1.41, 1.66, 1.90, 2.18])
f=np.array([1.4, 1.65, 3.0, 3.95, 4.3, 5.20, 6.85, 7.4])
df=np.array([ 0.4, 0.5, 0.4, 0.5, 0.6, 0.5, 0.5, 0.4])
bestfit = np.polyfit(d,f,1)
bestfitx = np.linspace(0,2.5,num=50)
bestfity = bestfit[0]*bestfitx + bestfit[1]

fig,ax4 = plt.subplots(1,1)
ax4.errorbar(d,f, yerr=df,fmt='yd', ecolor='red',label= 'Data')
ax4.set(xlabel='distance(cm)',ylabel='force(N)')
ax4.plot(bestfitx, bestfity)
```
