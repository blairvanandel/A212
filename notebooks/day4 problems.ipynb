```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline ######finally figured out what this does!!
```

```python
x1 = np.arange(-1,3.01,0.01)
y1 = 3*x1**2

plt.figure()
plt.plot(x1,y1)
plt.xlim(-1,3)
plt.xlabel('x axis')
plt.ylabel('f(x)')
plt.title('y as a function of 3*x^2')
```

```python
x2 = np.arange(-15,15.01,0.01)
y2 = np.cos(x2)/(1+((1/5)*(x2**2)))
plt.figure()
plt.plot(x2,y2, color = 'r')
plt.xlim(-15,15)
plt.axhline(color='k')
plt.axvline(color='k')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('y as a function of cos(x)/(1+(x/5)^2)')
```
